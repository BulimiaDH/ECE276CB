```python
import os
path = "./new_result/dict_values([0.1, 180, 40, 40])dict_values([1, 0.1, 0.05, 3, 0.1, 0.95, 'PI'])/"
fig, axes = plt.subplots(ncols=5, nrows = 2 ,figsize=(10,4),sharex=True, sharey=True)
index = [1,3,5,7,9]
noises = [0.01,0.05,0.1,1]

for i in range(2):
    if i == 0:
        path = "./new_result/dict_values([0.1, 180, 40, 40])dict_values([1, 0.1, 0.01, 3, 0.1, 0.95, 'VI'])/"
    else:
        path = "./new_result/dict_values([0.1, 180, 40, 40])dict_values([1, 0.1, 0.01, 3, 0.1, 0.95, 'PI'])/"
    for j in range(5):
        if not os.path.exists(path + 'value' + str(index[j]).zfill(3) + '.npy'):
            continue
        x = np.load(path + 'value' + str(index[j]).zfill(3) + '.npy')
        im = axes[i][j].imshow(x,aspect='auto')
        axes[i][j].set_title(f'episode:{index[j]},noise:{0.1}',fontsize=10)
        fig.colorbar(im,ax =axes[i][j])
fig.suptitle('Comparison between VI and PI')
fig.savefig('VP_compare.png')
```

```python
noises = [0.01,0.1,0.5,2,3,4,5]
fig, axes = plt.subplots(ncols=len(noises), nrows = 2 ,figsize=(10,4),sharex=True, sharey=True)
for i in range(len(noises)):
    path = "./new_result/dict_values([0.1, 180, 40, 40])dict_values([1, 0.1, " + str(noises[i])+", 3, 0.1, 0.95, 'VI'])/"
    _,_, policy, value, _, _ = pickle.load(open(path + '/data.pickle', 'rb'))
    
    im = axes[0][i].imshow(policy,aspect='auto')
    axes[0][i].set_title(f'policy,noise:{noises[i]}',fontsize=10)
    cbar = fig.colorbar(im,ax =axes[0][i])
    cbar.ax.tick_params(labelsize=7)
    im = axes[1][i].imshow(value,aspect='auto')
    axes[1][i].set_title(f'value,noise:{noises[i]}',fontsize=10)
    cbar = fig.colorbar(im,ax =axes[1][i])
    cbar.ax.tick_params(labelsize=7)
fig.savefig('./noise.png')
```
