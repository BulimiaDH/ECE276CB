```python
import os
import pickle
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.use('Qt4Agg')
from mpl_toolkits.mplot3d import Axes3D
plt.ion()
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import seaborn
%matplotlib qt
import pandas as pd
```

```python
pd_data = pd.DataFrame()
for i,path in enumerate(os.listdir('../result')):
    path = '../result/'+path
    data = pickle.load(open(path + '/data.pickle', 'rb'))
    traj = pickle.load(open(path + '/avg_len.pickle', 'rb'))
    _, hyperparameters, Q, policy = data
    hyperparameters['traj'] = traj[1]
    hyperparameters['path'] = path
    tmp_data = pd.DataFrame(hyperparameters,index = [i])
    pd_data = pd_data.append(tmp_data)
    
```
